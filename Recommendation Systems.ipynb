```python
del df1['Timestamp']
del df1['Age?']
del df1['Gender']
del df1['Preferred Genres']
del df1['ID']

```

```python
df1.fillna(0, inplace = True)
df1.head()
```

```python
df1.head()
```

```python
ratings = df1.as_matrix()
print ratings
```

```python
def similarity_calculation(ratings):
    sim = ratings.T.dot(ratings) 
    norms = np.array([np.sqrt(np.diagonal(sim))])
    return (sim / norms / norms.T)
```

```python
item_similarity = similarity_calculation(ratings)
```

```python
test = pd.read_csv('test.csv')
```

```python
test.head()
```

```python
test1 = test.copy()
```

```python
del test1['Age?']
del test1['Gender']
del test1['Preferred Genres']
del test1['ID']

```

```python
test1.fillna(0, inplace = True)
test1.head()
```

```python

ratings_test = test1.as_matrix()



print ratings_test
```

```python
def predict_topk(ratings, similarity, k=3):
    pred = np.zeros(ratings.shape)
    for j in xrange(ratings.shape[1]):
        top_k_items = [np.argsort(similarity[:,j])[:-k-1:-1]]
        for i in xrange(ratings.shape[0]):
            pred[i, j] = similarity[j, :][top_k_items].dot(ratings[i, :][top_k_items].T) 
            pred[i, j] /= np.sum(np.abs(similarity[j, :][top_k_items]))        
    
    return pred
```

```python
pre = predict_topk(ratings_test, item_similarity, k = 3)
```

```python

print pre
```

```python
df_revised = pd.DataFrame(data = pre, columns = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p'])
```
