Import pandas and numpy libraries.

```
import pandas as pd
import numpy as np
```

Import jointsx01 dataset into pandas dataframe.


```
jointsx01_downloaded = drive.CreateFile({'id': '1lFcHcuYTIv4zZbh2UhBEFWAlJq9XxGSN'})
jointsx01_downloaded.GetContentFile('jointsx01.csv')
```

```
df_jointsx01 = pd.read_csv('jointsx01.csv')
df_jointsx01.rename(index=str, columns={"ID": "id"}, inplace = True)
df_jointsx01.head()
```

Import outcomes99 dataset into pandas dataframe

```
outcomes99_downloaded = drive.CreateFile({'id': '1fGKcma3Nwk0ZxQ0XBYb9v7C19jrZgtal'})
outcomes99_downloaded.GetContentFile('outcomes99.csv')
```

```
df_outcomes99 = pd.read_csv('outcomes99.csv')
df_outcomes99.head()
```

Pick participants with confirmed right knee replacements or those who have self reported right knee replacements


```
#pick participants with confirmed right knee replacements or those who have self reported right knee replacements 

right_knee_replacements = df_outcomes99.loc[(df_outcomes99['V99ERKRPCF'] == 1) | (df_outcomes99['V99ERKRPCF'] == 2) | (df_outcomes99['V99ERKRPCF'] == 3) ]
right_knee_replacements.head()
```

Since, it is a binary classification problem replace the labels with knee replacements with 1 denoting past right knee replacement. The NaN are replaced with 0.


```
df_outcomes99['V99ERKRPCF'] = df_outcomes99['V99ERKRPCF'].replace([1,2,3],1)
df_outcomes99['V99ERKRPCF'] = df_outcomes99['V99ERKRPCF'].fillna(0)
df_outcomes99.head()
```

Filter only the id and right knee replacement columns from the outcomes dataframe.

```
right_knee_replacements_outcome = df_outcomes99.filter(['id','V99ERKRPCF'], axis=1)
right_knee_replacements_outcome.head()
```

Merge the jointx01 dataframe with right knee replacements outcome dataset. The id is used as the key.

```
joints_merged_right_knee = df_jointsx01.merge(right_knee_replacements_outcome, 'left', 'id')
joints_merged_right_knee.head()
```

The columns id and version are dropped as they are not required for training.


```
joints_merged_right_knee.drop(columns=['id', 'VERSION'], axis =1, inplace = True)
```

```
joints_merged_right_knee.head()
```

Define x and y for training. 

```
#drop the label column which is V99ERKRPCF
X = joints_merged_right_knee.drop(columns = 'V99ERKRPCF')
#fill NaN with 0s
X.fillna(0, inplace = True)
#define y 
y = joints_merged_right_knee['V99ERKRPCF']
```

Do a train test split 

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.3, random_state=42)
```

```
X_train.shape
```

Use logistic regression to train model.

```
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(X_train, y_train)

```

Predict test data

```
pred = lg.predict(X_test)
```

Check accuracy of model.


```
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)


```
